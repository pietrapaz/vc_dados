```python
import cv2 as cv
import time
```

```python
cap = cv.VideoCapture('teste.mp4')

if cap.isOpened == False:
    print("Erro")

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret == True:
        time.sleep(1/20)
        cv.imshow('frame', frame)
        if cv.waitKey(1) & 0xFF == ord('q'):
            break;
    else:
        break;

cap.release()
cv.destroyAllWindows()
```
